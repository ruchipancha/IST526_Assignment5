# Loading Data

```
df_aapl = pd.read_csv('data/AAPL.csv', parse_dates=True, index_col = "Date")


# adding Date as column and index
df_aapl['Date'] = df_aapl.index

df_ = df_aapl
# adding daily interval
df_.index = pd.DatetimeIndex(df_.index).to_period('D')

# display
display(df_)
```

# Time Series with Range

```
fig = go.Figure(go.Scatter(    
    x=df_['Date'], 
    y=df_['Open'],
    name = 'Open'
))
fig.update_layout(
    title = "Stock Price Open",
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=3,
                     label="3 months",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6 months",
                     step="month",
                     stepmode="backward"),
                dict(count=9,
                     label="9 months",
                     step="month",
                     stepmode="todate"),
                dict(count=1,
                     label="1 year",
                     step="year",
                     stepmode="backward"),
                dict(count=3,
                     label="2020-2022",
                     step="year",
                     stepmode="todate"),
                dict(count=10,
                     label="10 years",
                     step="year",
                     stepmode="todate"),
                dict(count=1,
                     label="Latest",
                     step="day",
                     stepmode="backward"),
                dict(step="all",
                    label="All"),
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)
fig.show()
```

Note: Plotly express was not correctly displaying the trace inside the range slider. I found a stack overflow exchange where they mentioned that for a size > 500, range slider trace may not be visible. This is why I chose to plot my graphs using Go.

https://stackoverflow.com/questions/67671431/plotly-range-slider-not-being-displayed-for-row-count-500

# Moving Average

```
df_rolling = df_['Open'].rolling(window = 5).mean()
df_rolling = df_rolling.fillna(df_rolling.mean())

fig.add_trace(
    go.Scatter(
          x = df_['Date'], 
          y = df_rolling, 
          name="Moving Average",
          line_width = 1
        )
    )

fig.show()
```

Moving average for 5 days looks very similar to the actual plot.

I am also plotting a moving average for 30 days to gain a better understanding of how the average looks monthly

```
df_rolling1 = df_['Open'].rolling(window = 30).mean()
df_rolling1 = df_rolling1.fillna(df_rolling1.mean())

fig.add_trace(
    go.Scatter(
          x = df_['Date'], 
          y = df_rolling1, 
          name="Moving Average 30 days",
          line_width = 1
        )
    )

fig.show()
```
