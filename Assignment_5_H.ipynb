```python
df = pd.read_csv('data/news_articles.csv', encoding='latin-1')
display(df.head())
display(df.tail())
```

# Data Cleaning

```python
for idx, row in df.iterrows():
  if '>' in row['Article']:
    row['Article'] = row['Article'].split('>')[1]
display(df)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
display(df['NewsType'].unique())
```

Splitting the data by news type

```python
df_sports = df[df['NewsType'] == 'sports']
display(df_sports)
```

```python
df_business = df[df['NewsType'] == 'business']
display(df_business)
```

Grouping data by week

```python
df_g_business = pd.DataFrame(df_business.groupby(pd.Grouper(key='Date', axis=0, freq='W'),as_index = False).size())
display(df_g_business)
```

```python
df_g_sports = pd.DataFrame(df_sports.groupby(pd.Grouper(key='Date', axis=0, freq='W'),as_index = False).size())
display(df_g_sports)
```

#Plotting time-series graph for business and sports news

```python
fig_full = go.Figure(
    go.Scatter(
        x = df_g_business['Date'],
        y = df_g_business['size'],
        name = 'Business'
    )
)
fig_full.add_trace(
    go.Scatter(
        x = df_g_sports['Date'],
        y = df_g_sports['size'],
        name = 'Sports'
    )
)

fig_full.update_layout(
    title="News Articles",
    yaxis_title="Number of Articles",
    xaxis = dict(
        title = 'Weeks',
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)
fig_full.show()
```
